```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("/content/car_crashes.csv")
plt.scatter(data['speeding'], data['alcohol'],)

plt.title("Scatter plot")

plt.xlabel('speeding')
plt.ylabel('alcohol')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("/content/car_crashes.csv")
plt.bar(data['speeding'], data['alcohol'], width=0.2)

plt.title("Bar Chart")

plt.xlabel('speeding')
plt.ylabel('alcohol')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("/content/car_crashes.csv")

sns.histplot(x='speeding', data=data,)
plt.title("Histogram")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("/content/car_crashes.csv")
plt.boxplot(data['speeding'], data['alcohol'], data['not_distracted'])

plt.title("Box Plot")

plt.xlabel('speeding')
plt.ylabel('total')

plt.show()
```
